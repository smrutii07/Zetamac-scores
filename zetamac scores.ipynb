```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
scores = pd.read_csv('Downloads/zetamac/Zetamac scores.csv')
```

```python
scores.columns
```

```python
scores = scores.drop('Unnamed: 0', axis = 1)
scores = scores.astype('int64')
```

```python
plt.figure(figsize=(15,9))
plt.xlabel('# tries', fontsize = 20)  
plt.ylabel('scores',fontsize = 20)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.gca().xaxis.set_major_locator(MultipleLocator(5))

num_lines = scores.shape[1]
custom_colors = ['darkmagenta', 'darkgreen', 'darkblue']

#colors = plt.cm.viridis(np.linspace(0, 1, num_lines))

for i, column in enumerate(scores.columns):
    plt.plot(scores.index, scores[column], color=custom_colors[i], label=column)
    
plt.savefig('Downloads/zetamac/scores_plot.png')
    
plt.legend(fontsize=15)
```
